```python
import numpy as np
import math
print("input point like [x y]")
edgestart = np.array([float(n) for n in input("start point of this edge: ").split()])
edgeend = np.array([float(n) for n in input("end point of this edge: ").split()])
roundcenter = np.array([float(n) for n in input("center point of this arc:").split()])

while np.linalg.norm(edgestart - roundcenter, 2) != np.linalg.norm(edgeend - roundcenter, 2):
    print("wrong center")
    print("edgestart --> roundcenter", np.linalg.norm(edgestart - roundcenter, 2))
    print(" edgeend --> roundcenter", np.linalg.norm(edgeend - roundcenter, 2))
    roundcenter = np.array([float(n) for n in input("center point of this arc:").split()])
# while edgeend[0] <= edgestart[0]:
#     print("wrong end")
#     roundcenter = np.array([float(n) for n in input("center point of this arc:").split()])
radius = np.linalg.norm(edgestart - roundcenter, 2)
```

```python
thetarange = math.acos((2 * radius ** 2 - np.linalg.norm(edgeend - edgestart, 2) ** 2)/(2 * radius ** 2))
num = 90
shape = []
for i in range(num):
    theta = i * thetarange / num
    tempx = radius * math.cos(- theta + 
              math.atan2(edgestart[1] - roundcenter[1], edgestart[0] - roundcenter[0])) + roundcenter[0]
    tempy = radius * math.sin(- theta +
              math.atan2(edgestart[1] - roundcenter[1], edgestart[0] - roundcenter[0])) + roundcenter[1]
    shape.append([tempx,tempy])
for i in range(num):
    print('{0},{1}'.format(shape[i][0],shape[i][1]),end=" ")
```
